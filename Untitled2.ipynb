```python
%matplotlib inline
import numpy as np

size = 3
values = [[100,100,100],[100,100,100],[100,100,0]]
goal = (2,2)
actions = ["up","down","left","right"]
policy = [["up","up","up"],["up","up","up"],["up","up",None]]
```

```python
import matplotlib.pyplot as plt  
import matplotlib.patches as patches

def draw(mark_pos, action):
    fig, ax = plt.subplots()
    mp = ax.pcolor(values, cmap=plt.cm.YlOrRd,vmin=0,vmax=8)
    ax.set_aspect(1)
    ax.set_xticks(range(size), minor=False)
    ax.set_yticks(range(size), minor=False)

    for x in range(len(values)):
        for y in range(len(values[0])):
            plt.text(x+0.5,y+0.5,int(1000*values[x][y])/1000,ha = 'center', va = 'center', size=20)

    plt.text(goal[0]+0.75,goal[1]+0.75,"G",ha = 'center', va = 'center', size=20)
    
    if mark_pos == "all":   # 指定した位置にactionの文字列を書くという処理
        for x in range(size):
            for y in range(size):
                plt.text(x+0.5,y+0.25,policy[x][y],ha = 'center', va = 'center', size=20)
    elif mark_pos != None:
        plt.text(mark_pos[0]+0.5,mark_pos[1]+0.25,action,ha = 'center', va = 'center', size=20)
        
    plt.show()
    fig.clear()
            
draw(None,None)
```

```python
def postvalue(pos,action):
    p = [pos[0],pos[1]]
    if   action == "up":    p[1] += 1
    elif action == "down":  p[1] -= 1
    elif action == "left":  p[0] -= 1
    elif action == "right": p[0] += 1
        
    for i in [0,1]:  # 壁にぶつかったらそのまま
        if p[i] < 0:     p[i] = 0
        if p[i] >= size: p[i] = size - 1
            
    return values[p[0]][p[1]]

def action_value(pos, action, goal):
    if pos == goal: return values[pos[0]][pos[1]]    # ゴールの価値は不変
    
    cur_v = values[pos[0]][pos[1]]
    post_v = postvalue(pos,action)
    
    water = 0.0
    if pos == (1,2): water += 10
    
    return 1.0 + 0.9*post_v + 0.1*cur_v + water
```

```python
action_value((2,1),"up",goal)
```

```python
action_value((0,0),"down",goal)
```

```python
import math

def sweep():
    changed = False
    
    for x in range(size):
        for y in range(size):
            best_cost = 1000000
            best_action = None
            
            for a in actions:
                q = action_value((x,y), a, goal)
                if q < best_cost:
                    best_cost = q
                    best_action = a
                    
            if math.fabs(values[x][y] - best_cost) > 0.0001:
                changed = True
                values[x][y] = best_cost
                policy[x][y] = best_action
                draw((x,y),best_action)
                
    return changed
```

```python
draw(None,None)
```

```python
changed = True
```

```python
n = 1
while changed:
    print("sweep " + str(n))
    changed = sweep()
    n += 1
```
